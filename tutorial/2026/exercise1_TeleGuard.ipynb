# Welcome to the Puma Tutorial

Puma — *Programmable Utility for Mobile Automation* — is a Python library that lets you control Android apps through simple, high‑level actions.
It supports a variety of popular applications such as **Telegram**, **TeleGuard**, **WhatsApp**, **Google Maps**, and more (see the README.md)

In this tutorial you will learn how to use Puma to interact with the a few of these apps: Sending messages with TeleGuard, navigating using Google Maps and conduct web searches using Google Chrome. Finally, you will build a full scenario, using multiple applications simultaneously to mimic real user behavior.

The workshop is primarily aimed at _using_ Puma. If you’re interested in extending Puma itself by extending or improving features, or by adding support for new apps, check out bonus **Exercise 5**. ### Appium
Appium is a framework for testing applications, based on Selenium. In Puma, we use Appium to execute actions on the device.
In this context, we define functions in Python that exist of one or more `Appium actions`. An Appium action is an action on the phone, for instance tapping on an element. In our Python code, we define functions executing one or more
Appium actions, for instance `send_message()` where the message box is tapped, text is added to the box and the send button is clicked.

## 3. Send a picture with TeleGuard
Apart from sending messages, Puma also supports sending pictures. Each application has a `README.md`, listing all functionality of an application and the options. Sending a picture can be done in several ways, e.g. sending a picture from the gallery or taking a new picture with the camera. Lastly, an optional caption can be added.
_Note_:
`picture_id` is the index of the picture shown in the Android media picker (the first picture has index 1).
If `picture_id` is omitted, TeleGuard will open the camera, take a picture, and send it.
`caption` is optional – you can omit it to send the image without text.

````
teleguard.send_picture(conversation="",
                       #optional parameters, you can also remove them
                       picture_id=1, caption="")
````

#### What to observe
After running the cell, the TeleGuard app on the device should open the media picker (or camera) and automatically send the selected image.
The caption (if provided) appears in the chat bubble alongside the picture.
If you omitted `picture_id`, a new photo will be captured and sent.
Feel free to experiment with different picture_id values or captions to see how the app behaves.
